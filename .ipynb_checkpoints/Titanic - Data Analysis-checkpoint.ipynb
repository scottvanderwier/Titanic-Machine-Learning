```python
sns.factorplot('Sex',data=titanic_df,hue='Pclass',kind='count')
```

```python
sns.factorplot('Pclass',data=titanic_df,hue='Sex',kind='count')
```

```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
titanic_df[0:10]
```

```python
sns.factorplot('Pclass',data=titanic_df,hue='person',kind='count')
```

```python
fig = sns.FacetGrid(titanic_df,hue='Sex',aspect=2)
fig.map(sns.kdeplot,'Age',shade=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df,hue='person',aspect=2)
fig.map(sns.kdeplot,'Age',shade=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df,hue='Pclass',aspect=2)
fig.map(sns.kdeplot,'Age',shade=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = list()

for level in deck:
    levels.append(level[0])
    
levels.sort()
    
cabin_df = DataFrame(levels)

cabin_df.columns = ['Cabin']

sns.factorplot('Cabin',data=cabin_df,kind='count',palette='winter_d')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']

sns.factorplot('Cabin',data=cabin_df,kind='count',palette='summer')
```
